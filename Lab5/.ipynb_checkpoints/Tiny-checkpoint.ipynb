```python
import cv2
import math
import numpy as np
from datetime import datetime
from utilities import get_data, plot_heatmap, plot_res
```

```python
train_x, train_y, test_x, test_y = get_data(gray=True, size=16)
train_x = np.array([x.flatten() for x in train_x])
test_x = np.array([x.flatten() for x in test_x])
```

```python
print('=====KNN with cv2.ml.KNearest function=====')


knn = cv2.ml.KNearest_create()
knn.train(train_x, cv2.ml.ROW_SAMPLE, train_y)

for k in range(5):
	start = datetime.now()
	ret, results, neighbours ,dist = knn.findNearest(test_x, k+1)

	count = 0
	hit = 0
	for i in range(len(test_x)):
		if results[i] == test_y[i]:
			hit += 1
		count += 1
	end = datetime.now()
	print('accu with ', k+1, ' neighbors = ', hit/count)
	print('duration time = ', end - start)


print('=====KNN with Manhattan diatance=====')

for k in range(5):
	start = datetime.now()
	results_self = []
	for i in range(len(test_x)):
		store = []
		for j in range(len(train_x)):
			store.append(sum(abs(test_x[i] - train_x[j])))
		ans_idxs = np.argsort(store)[0:k+1]
		final_idx = np.bincount(ans_idxs).argmax()
		results_self.append(train_y[final_idx])
		#print('results_self : ', results_self)

	count = 0
	hit = 0
	for i in range(len(test_x)):
		if results_self[i] == test_y[i]:
			hit += 1
		count += 1
	end = datetime.now()
	print('accu  with ', k+1, 'neighbors = ', hit/count)
	print('duration time = ', end - start)
```
